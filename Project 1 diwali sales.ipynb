```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # vizulization
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding="unicode_escape")
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(["Status","unnamed1"],axis=1,inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)    #drop null values
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

```python
df.columns
```

```python
df.describe()
```

```python
plt.figure(figsize=(5,3))
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)


```

```python
plt.figure(figsize=(5,3))
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```


*From above graphs we can see that most of the buyers are female even the purchasing power of female is greater than men*

```python
plt.figure(figsize=(5,3))
ax=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
plt.figure(figsize=(5,3))
sales_gen=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set_palette("Paired")
sns.barplot(x='Age Group',y='Amount',data=sales_gen)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

```python
df.columns
```

```python
plt.figure(figsize=(16,5))
sales_gen=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.set_palette("Paired")
sns.barplot(x='State',y='Orders',data=sales_gen)
```

```python
plt.figure(figsize=(16,5))
sales_gen=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set_palette("Paired")
sns.barplot(x='State',y='Amount',data=sales_gen)
```

```python
ax=sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
sns.set_palette("pastel")
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
plt.figure(figsize=(7,5))
sale_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set_palette("pastel")
sns.barplot(x='Marital_Status',y='Amount',data=sale_state,hue="Gender")
```

*From above Graphs we can see that most of the buyers are married(women) and they have high purchasing power* 

```python
ax=sns.countplot(data=df,x='Occupation')
sns.set(rc={'figure.figsize':(20,5)})
sns.set_palette("pastel")
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
plt.figure(figsize=(19,8))
sale_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set_palette("pastel")
sns.barplot(x='Occupation',y='Amount',data=sale_state)
```

*From above graphs we can see that most of the buyers are working in IT,Healthcare and Aviation Sector*

```python
ax=sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(24,10)})
sns.color_palette("coolwarm",as_cmap=True)
for bars in ax.containers:
 ax.bar_label(bars)
```

```python
plt.figure(figsize=(19,8))
sale_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(10)
sns.set_palette("husl")
sns.barplot(x='Product_Category',y='Amount',data=sale_state)
```

*From above graphs we can see that most of the sold products are from Food,Clothing and Electronics category*

```python
plt.figure(figsize=(19,8))
sale_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set_palette("Set2")
sns.barplot(x='Product_ID',y='Orders',data=sale_state)
```

# Conclusion:

*Married Group Age group 26-35 yrs from UP,Maharastra and Karnataka working in IT,Healthcare and Aviation are more likely to buy products from Food,Clothing and Electronics category*
